## 4.1 Information inside a computer

When using a computer everyday, it's important to be able to extract useful information from the computer. For instance, reading emails requires being able to access the information in your inbox. 

Unlike reading emails, quantum computing works at the lowest level of the system. It would be like directly accessing the bits in a classical computer.  

Instead of working with bits, quantum computing works with qubits. To do anything at all with qubits, it's important to describe what state they're in. The state of a qubit is described by something known as a ket. 


### 4.1.1 Ket?


How do we represent a quantum state? It is easy to think this would require a very complicated mechanism. It is as simple as a column vector! The state of a quantum system can be described with a vector known as a ket. Usually, the state of a system is given by the ket $\ket{\psi}$.

The coins still work as an example to illustrate the state of the system. If the coin is heads up we can use the ket $\ket{H}$, if it's tails up it is $\ket{T}$.

When the coin is spinning in the air, the superposition state (see section 1.3), this state can be represented by the state $\ket{\psi}$: 

$$
\ket{\psi} = \frac{1}{\sqrt{2}}[\ket{H} + \ket{T}]
$$

Since the coin has a $\frac{1}{2}$ probability of being in either heads or tails, one may have intuitively expected this state to have a $\frac{1}{2}$ instead of $\frac{1}{\sqrt{2}}$. The reason for this is that in quantum mechanics, a state is described with an amplitude (the factor by which it is multiplied) proportional to the square root of the probability. 

It can be convenient to define the heads state using a ket as

$$
\ket{H} = 

\begin{bmatrix}
1 \\ 0
\end{bmatrix}
$$

Similarly, the tails side up can be represented by the ket

$$
\ket{T} = 

\begin{bmatrix}
0 \\ 1
\end{bmatrix}
$$

This allows $\ket{\psi} $ to be expanded as 

$$
\ket{\psi} = \frac{1}{\sqrt{2}}\left(\begin{bmatrix}
1 \\ 0
\end{bmatrix}+ \begin{bmatrix}
0 \\ 1
\end{bmatrix}\right)
$$

By adding the corresponding elements of each vector, this can be simplified to

$$ \ket{\psi} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix} $$

Instead, if there were a biased coin which had a probability of heads given by $P$, the probability of tails would then be $1-P$.  The state of the spinning coin could be represented as:

$$
\ket{\psi} = \sqrt{P}\ket{H} + \sqrt{1-P}\ket{T}
$$

Or 

$$
\ket{\psi} = \begin{bmatrix}
\sqrt{P} \\ \sqrt{1-P}
\end{bmatrix}
$$

### 4.1.2 Bra?

Where the bra $\bra{H}$ is the same as the ket $\ket{H} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ but a row vector where the complex numbers are all multiplied by $-1$. Since there are no imaginary numbers in $\ket{H}$, the bra $\bra{H}$ is just the transpose of the column vector as

$$

\bra{H} = [\ket{H}^*]^T = \begin{bmatrix} 1 & 0 \end{bmatrix}
$$

More generally, the state of a quantum system $\ket{\psi}$ is referred to as a *wavefunction*. The wavefunction describes all the information we have availible about a system. In quantum computing the state of the quantum computer is described by the wavefunction (in later chapters a more advanced formalism will be introduced based off the wavefunction).

## 4.2 Probability of measurement



For the coins, one may wish to know the probability of flipping the coin and getting heads or tails up. With Dirac notaion, the probability of an outcome (heads up) can be calculated using the inner product. This is done as



Using the coin example, the states that the probability of the coin ending up in the heads state $\ket{H}$ can be calculated as

$$ P(H) = |\braket{\psi|H}|^2  $$

Explicity computing this gives


$$
P(H) = |\braket{\psi|H}|^2 =   \Bigg |\frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} \Bigg |^2
$$

$$ = \left| \frac{1}{\sqrt{2}} \right|^2 = \frac{1}{2} $$

Since all the probabilities of the coin landing on $\ket{H}$ or $\ket{T}$ must add to 1, for a biased or unbiased coin, it must be that:

$$ |\braket{\psi|\psi}|^2 = 1 $$

This can be easily shown with the previous definition of $\ket{\psi}$

$$
\Bigg |\begin{bmatrix} \sqrt{P} \\ \sqrt{1-P} \end{bmatrix} \begin{bmatrix} \sqrt{P} & \sqrt{1-P} \end{bmatrix} \Bigg |^2
$$



$$ =  |\sqrt{P}\sqrt{P} + \sqrt{1-P}\sqrt{1-P}|^2 $$

$$ =  |P + 1-P|^2 = 1 $$
